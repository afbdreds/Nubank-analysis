```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()
df_ = pd.DataFrame()
files = os.listdir('.')
for file in files:
    if "csv" in file:
        df_ = pd.read_csv(file)
    df = pd.concat([df, df_])
```

```python
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
# Group by 'date' and 'category' and sum the 'amount'
grouped_category = df.groupby(['category']).agg({'amount': 'sum'}).sort_values("amount",ascending=False).reset_index()

grouped_category.head()
```

```python
grouped_establishment = df.groupby(['title','category']).agg({'amount': 'sum'}).sort_values("amount",ascending=False).reset_index()

grouped_establishment.head(10)
```
